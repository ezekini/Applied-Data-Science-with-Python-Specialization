# Research question: Is there a correlation between the performance of the best state NBA team (Detroit Pistons) and the U. of Michigan basketball team (Michigan Wolverines)? 70' onwards

# Dataframe 1

***First dataset: Historic data for the Michigan Wolverines basketball team.***

- Conf: Conference
- W: Wins
- L: Losses
- W-L%: Win-Loss percentage
- W.1: Conference wins
- L.1: Conference loss
- W-L%.1: Conference Win-Loss percentage	
- SRS: Simple Rating System.A rating that takes into account average point differential and strength of schedule. The rating is denominated in points above/below average, where zero is average. Non-Division I games are excluded from the ratings.
- SOS: Strength of Schedule. A rating of strength of schedule. The rating is denominated in points above/below average, where zero is average. Non-Division I games are excluded from the ratings.
- PTS: Points Per Game
- PTS.1: Opponent Points Per Game 
- AP Pre: Rank in preseason poll
- AP High: Highest rank in AP poll during season
- AP Final: Rank in final AP poll (poll taken before the conclusion of the NCAA tournament)
- NCAA Tournament 	
- Seed 	
- Coach(es)

_URL_ https://www.sports-reference.com/cbb/schools/michigan/

```python
df1 = pd.read_csv('michigan-basketball-hist.csv', header=1)
```

```python
df1
```

```python
df1.info()
```

```python
# Drop useless cols
df1.drop(labels=['Rk', 'Conf', 'W.1', 'L.1', 'W-L%.1', 'AP Pre', 'AP High', 'AP Final',
                 'Seed', 'SRS', 'SOS', 'NCAA Tournament', 'Coach(es)'], axis=1, inplace=True)
```

```python
#There are some * in ['W', 'L', 'W.1', 'L.1'] that we need to remove in order to change the datatype
df1[['W', 'L']] = df1[['W', 'L']].apply(lambda x: x.str.replace('*',''))
```

```python
df1['Season'] = df1['Season'].apply(lambda x: x[:4]) 
```

```python
#Now let's change datatypes
df1[['Season','W', 'L']] = df1[['Season','W', 'L']].apply(pd.to_numeric) 
```

```python
df1.info()
```

```python
df1 = df1.loc[(df1['Season'] >= 1970) & (df1['Season'] < 2019)]
```

```python
df1 = df1.sort_values(by='Season')
```

```python
df1.reset_index(inplace=True, drop=True)
```

```python
df1
```

# Dataframe 2

***2nd dataset: Historic data for the Detroit Pistons (Michigan) NBA basketball team.***

- Conference: An athletic conference is a collection of sports teams, playing competitively against each other at the professional, collegiate, or high school level. In many cases conferences are subdivided into smaller divisions, with the best teams competing at successively higher levels. Conferences often, but not always, include teams from a common geographic region.
- Win %: In sports, a winning percentage is the fraction of games or matches a team or individual has won. It is defined as wins divided by the total number of matches played. A draw counts as a 1⁄2 win.
- GB: In most North American sports, the phrase games behind or games back is a common way to reflect the gap between a leading team and another team in a sports league, conference, or division.

```python
url2 = 'https://en.wikipedia.org/wiki/List_of_Detroit_Pistons_seasons'
```

```python
df2 = pd.read_html(url2, header=0)
```

```python
df2[1]
```

```python
#Let's get the rows we need
df2 = df2[1].loc[24:].drop(labels=[73,74],axis=0)
```

```python
df2.reset_index(inplace=True, drop=True)
```

```python
df2.columns
```

```python
# Let's drop columns and rename a few
df2.drop(labels=['Team Season', 'Conference', 'Finish', 'Division',
                 'Finish.1', 'Playoffs', 'Awards'], axis=1, inplace=True)
```

```python
df2.rename(columns={'NBA Season': 'Season', 'Wins':'W', 'Losses':'L', 'Win%':'W-L%'}, inplace=True)
```

```python
df2.info()
```

```python
df2['Season'] = df2['Season'].apply(lambda x: x[:4])
```

```python
#Now let's change datatypes
df2[['Season', 'W', 'L', 'W-L%']] = df2[['Season', 'W', 'L', 'W-L%']].apply(pd.to_numeric) 
```

```python
df2.info()
```

```python
df2
```

---

Let's create a new columnd with a window of 4 years (excluding current). Since the Wolverines is a uni team, we can assume that good players may end up in the NBA after a few years. 

```python
df1['MA'] = df1['W-L%'].rolling(window=4).mean().shift(1)
```

# Plot

```python
sns.set(style='whitegrid')
```

```python
import matplotlib.ticker as ticker
fig = plt.figure()

ax = fig.add_subplot(1, 1, 1)
ax.plot(df1['Season'], df1['W-L%'], 'o-', color='#37e17e', linewidth=0.8)
ax.plot(df2['Season'], df2['W-L%'], 'o-', color='#007db3', linewidth=0.8)
#ax.plot(df1['Season'], df1['MA'], '-', color='#cb6a49', alpha=0.5)


# Axes
ax = plt.gca()
ax.axis([1969, 2019, 0, 0.9])
plt.xticks([1970, 1980, 1990, 2000, 2010, 2020], fontsize=12)
plt.yticks([0.2, 0.4, 0.6, 0.8], fontsize=12)
plt.xlabel('Year', fontsize=15)
plt.ylabel('W-L%', fontsize=15)

# grid
plt.grid(b=True, which='both', axis='both', linestyle='--')
ax.set_axisbelow(True)

# Title
plt.title('Win-Loss % for the Michigan Wolverines & Detroit Pistons (1970-2018)',
          fontsize=20, pad=25)
# Legend
plt.legend(['Michigan Wolverines', 'Detroit Pistons'],
           fontsize=14, loc='lower right', frameon=True)
#plt.legend(['Michigan Wolverines', 'Detroit Pistons', 'Michigan Wolverines 4 year Moving Average'],
#           fontsize=14, loc='lower right', frameon=True)

# W-L% Text
props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)
text = r'$W-L\% = \frac{Wins}{Wins+Losses}$'
ax.text(0.04, 0.14, text, transform=ax.transAxes, fontsize=15,
        verticalalignment='top', bbox=props)


# Hide the right and top spines
ax.spines['bottom'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['top'].set_visible(False)
plt.savefig('data.png')
```
